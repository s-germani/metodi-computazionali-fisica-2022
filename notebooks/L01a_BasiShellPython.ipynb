### Liste

```python
# Lista Animali 
lista_animali = ['Cane', 'Gatto', 'Cavallo', 'Asino']

print(lista_animali)
```

```python
# Numero elementi in una lista
lunghezza = len(lista_animali)
print('Lunghezza lista:', lunghezza )
```

### Ciclo for

```python
ln = [1,3,5,7,9,11,22,33]

# ciclo for su lista ln
for i in ln:
    print(i)
```

```python
lcolori = ['rosso', 'verde', 'blu', 'giallo', 'viola']

# ciclo for su lista lcolori
for c in lcolori:
    print(c)
```

```python
# ciclo for su range (0-9) e calocolo somma

somma = 0
for j in range(10):
    somma = somma + j
    print(j)
    

print('somma', somma)
```

```python
for j in range(2,10):
    print(j)
    
```

```python
for j in range(10, 21, 2):
    print(j)
```

### Ciclo while

```python
check = 0

while check < 5:    
    check = check + 1
    print(check)
    
    
```

### Istruzioni condizionali (if-elif-else)

```python
avc = 10

if avc > 20:
    print(avc, '> 20')
    
```

```python
if avc < 20:
    print(avc, '< 20')
```

```python

bvc = 20
if bvc > 20:
    print('-->', bvc, '> 20')
elif bvc < 20:
    print('-->', bvc, '< 20')
else:
    print('-->', bvc, '= 20')
    
```

```python
for k in range(10):
    if k >=1 and k < 5: 
        print(k, '<<<')
    elif k == 5:
        print(k, '===')
    elif k> 5 and k < 9:
        print(k, '>>>')
    else:
        print(k, '***')
```

```python
for kk in range(0, 20):
    if kk%2 == 0 or kk == 19: 
        print(kk)
```
